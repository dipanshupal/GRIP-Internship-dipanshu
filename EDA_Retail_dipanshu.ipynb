```python
corelation=new_data.corr()
```

```python
sns.heatmap(corelation, xticklabels=corelation.columns, yticklabels=corelation.columns, annot=True)
```

```python
sns.pairplot(new_data)
```

```python
sns.relplot(x='Sales', y='Profit', hue='State', data=new_data)
```

```python
sns.relplot(x='Sales', y='Profit', hue='Region', data=new_data)
```

```python
sns.relplot(x='Sales', y='Profit', hue='Category', data=new_data)
```

```python
sns.relplot(x='Sales', y='Profit', hue='Discount', data=new_data)
```

```python
sns.distplot(new_data['Profit'], bins=20)
```

```python
sns.distplot(new_data['Discount'], bins=5)
```

```python
#visualising the category wise share in the sales
plt.figure(figsize=(6,6))
textprops= {"fontsize":15}
plt.title('Category')
plt.pie(new_data['Category'].value_counts(), labels=new_data['Category'].value_counts().index, autopct='%1.1f%%', textprops = textprops)
plt.show()
```

```python
#categiry wise share of sales and profit
plt.figure(figsize=(10,16))
new_data.groupby('Category')['Profit','Sales'].agg(['sum']).plot.bar()
plt.ylabel('Profit')
plt.show()
```

```python
#Category wise visualization of sales and profit
top_category_s = new_data.groupby('Category').Sales.sum().nlargest(n=100)
top_category_p = new_data.groupby('Category').Profit.sum().nlargest(n=100)
#visual plot
plt.style.use('seaborn')
top_category_s.plot(kind='bar', figsize=(10,5), fontsize = 14)
top_category_p.plot(kind='bar', figsize=(10,5), fontsize = 14, color='red')
plt.xlabel('Category', fontsize=15)
plt.ylabel('Total Sales/Profit', fontsize=15)
plt.title('Top category Sales v/s Profit', fontsize=15)
plt.show()

```

```python
# Sub-category wise visualization
top_subcategory_s = new_data.groupby("Sub-Category").Sales.sum().nlargest(n=100)
top_subcategory_p = new_data.groupby('Sub-Category').Profit.sum().nlargest(n=100)
#visual plot
plt.style.use('seaborn')
top_subcategory_s.plot(kind='bar', figsize=(10,5), fontsize = 14)
top_subcategory_p.plot(kind='bar', figsize=(10,5), fontsize = 14, color='red')
plt.xlabel('Sub-Category', fontsize=15)
plt.ylabel('Total Sales/Profit', fontsize=15)
plt.title('Top Sub-Category Sales v/s Profit', fontsize=15)
plt.show()
```

```python
#statewise visualization
plt.figure(figsize=(14,12))
statewise=new_data.groupby(['Sub-Category'])['Profit'].sum().nlargest(50)
statewise.plot.barh() #h for horizontal
```

```python
#profit vs discount
plt.figure(figsize=(8,7))
sns.lineplot(new_data['Discount'],new_data['Profit'], data=new_data )
```

```python
#region wise Profit visualization
plt.figure(figsize=(6,6))
plt.title('Region')
plt.pie(new_data['Region'].value_counts(), labels=new_data['Region'].value_counts().index, autopct='%1.1f%%')
plt.show()
```

```python
#sales and profit statewise for top 10 states
top_states_s = new_data.groupby("State").Sales.sum().nlargest(n=10)
top_states_p = new_data.groupby('State').Profit.sum().nlargest(n=10)
#visual plot
plt.style.use('seaborn')
top_states_s.plot(kind='bar', figsize=(10,5), fontsize = 14)
top_states_p.plot(kind='bar', figsize=(10,5), fontsize = 14, color='red')
plt.xlabel('States', fontsize=15)
plt.ylabel('Total Sales/Profit', fontsize=15)
plt.title('Top 10 states Sales v/s Profit', fontsize=15)
plt.show()
```
